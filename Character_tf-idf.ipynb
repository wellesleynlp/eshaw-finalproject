```python
import collections
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
import csv
import os

# list_of_characters = ['PUCK','EGEUS','HELENA','Pyramus','BOTTOM','HERMIA','FLUTE','QUINCE','THESEUS','TITANIA','LYSANDER',
# 'Prologue','Fairy','DEMETRIUS','Thisbe','HIPPOLYTA','PHILOSTRATE','OBERON']

list_of_characters=[]
token_dict = collections.defaultdict(list)

for subdir, dirs, files in os.walk('/home/eshaw/Documents/NLP/eshaw2-finalproject/csv'):
    print len(files)
    for file in files:
        list_of_characters.append(file[0:-4])
        file_path = subdir + os.path.sep + file
        shakes = open(file_path, 'r')
        text = shakes.read()
        token_dict[file] = text
```

```python
with open("dream.csv", "r") as corpus:
    reader = csv.reader(corpus, delimiter=',')
    word_corpus = reader.next()
    print len(word_corpus)

print len(files)
print len(set(word_corpus))
```

```python
tf = TfidfVectorizer(analyzer='word', min_df = 0)

tfidf_matrix =  tf.fit_transform(token_dict.values())
feature_names = tf.get_feature_names() 
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_dict = collections.defaultdict(list)
cosine_similarities = cosine_similarity(tfidf_matrix)

for c,character in enumerate(list_of_characters):
    cosine_similarities[c] = [val if val < cosine_similarities[c].max() else 0.0 for val in cosine_similarities[c]]
#     similarities_sorted = np.sort(cosine_similarities[c])[::-1]
    print cosine_similarities
#     similarity_dict[character].append(list_of_characters[np.argmax(cosine_similarities[c])])
# print similarity_dict
```
